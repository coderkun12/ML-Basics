#### 3. Function defination and usage

```python
def euc_norm(li):
    sum=0
    for i in li:
        sum+=(i**2)
    return sum**(1/2)
```

```python
print("Euclidian Norm for list [12,2,3,4] is: ",euc_norm([12,2,3,4]))
print("Euclidian Norm for list [12,24,56] is: ",euc_norm([12,24,56]))
```

#### 4. Find second largest element

```python
def find_sela(li):
    a=0
    b=0
    for i in li:
        if i>a:
            b=a
            a=i
        else:
            if b<i and i!=a:
                b=i
    return b

print("Second Largest element in list [10,2,20]: ",find_sela([10,2,20]))
print("Second largest element in list [3,5,2,9,7,4,9]: ",find_sela([3,5,2,9,7,4,9]))
```

#### 2. Create pairs and find ones lying inside the unit circle

```python
# To define pairs
def create_pairs(n):
    random_array=np.random.rand(n,2)
    random_array*=2
    random_array-=1
    return random_array
```

```python
# To get the unit pairs
pairs=create_pairs(100)
condition=np.logical_and(pairs[:,0]<1, pairs[:,0]>-1) & np.logical_and(pairs[:,1]<1, pairs[:,1]>-1)
indices=np.where(condition)
unit_pairs=pairs[indices]
```

```python
def find_points_inside_unit_circle(points):
  distances = np.sqrt(points[:, 0]**2 + points[:, 1]**2)
  return distances <= 1


points_inside = find_points_inside_unit_circle(pairs)
n1=0
for j in points_inside:
    if j==True:
        n1+=1

print(n1)
```

```python
# Calculation of pi using the monte-carlo method
pi_value=(n1/unit_pairs.shape[0])*4
pi_value
```

#### 3. Scatter plot of points inside and outside circle

```python
distances_from_origin = np.sqrt(unit_pairs[:, 0]**2 + unit_pairs[:, 1]**2)
inside_unit_circle = distances_from_origin <= 1

plt.figure(figsize=(8, 8))
plt.scatter(unit_pairs[inside_unit_circle, 0], unit_pairs[inside_unit_circle, 1], color='red', label='Inside')
plt.scatter(unit_pairs[~inside_unit_circle, 0], unit_pairs[~inside_unit_circle, 1], color='blue', label='Outside')

plt.title('Scatter Plot of Points Within and Outside the Unit Circle')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.gca().set_aspect('equal')

plt.show()
```

#### 4. Calculate center of mass

```python
def calculate_cm(masses,coordinates):
    x_cm_1=0
    y_cm_1=0
    z_cm_1=0
    for i in range(0,3):
        x_cm_1+=coordinates[i][0]*masses[i]
        y_cm_1+=coordinates[i][1]*masses[i]
        z_cm_1+=coordinates[i][2]*masses[i]
    x_cm=x_cm_1/sum(masses)
    y_cm=y_cm_1/sum(masses)
    z_cm=z_cm_1/sum(masses)
    return [x_cm,y_cm,z_cm]
masses=[0.5,1.3,2.7]
coordinates=[[0,0,0],[2.5,4,1.3],[8,-3,7.5]]
print("The center of mass of points with coordinates [[0,0,0],[2.5,4,1.3],[8,-3,7.5]] and respective masses [0.5,1.3,2.7] is: ",calculate_cm(masses,coordinates))
```

#### Visualization of the points

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with marker size proportional to mass
# Ensure masses and coordinates have the same length
masses = np.array(masses)  # Convert masses to a NumPy array

scatter = ax.scatter(*zip(*coordinates), s=masses * 100, alpha=0.7)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Visualization of Atoms')

# Show the plot
plt.show()
```
